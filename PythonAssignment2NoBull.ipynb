```python
# Import the Necessary Libraries

import requests
import pandas as pd
import json
```

```python
#Assign the api to a variable names 'url'

url= 'https://us-central1-bytehouse-prod-uk.cloudfunctions.net/collection?handle=mens-apparel%2Fdata--style--tops'
headers = {'accept': '*/*',
  'accept-encoding': 'gzip, deflate',
  'accept-language': 'en-US,en;q=0.9',
  'origin': 'https://uk.nobullproject.com',
  'referer': 'https://uk.nobullproject.com/',
  'sec-ch-ua': '"Not?A_Brand";v="8", "Chromium";v="108", "Google Chrome";v="108"',
  'sec-ch-ua-mobile': '?0',
  'sec-ch-ua-platform': '"Windows"',
  'sec-fetch-dest': 'empty',
  'sec-fetch-mode': 'cors',
  'sec-fetch-site': 'cross-site',
  'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
data='''handle=mens-apparel%2Fdata--style--tops'''
```

```python
#Store the response of the api using 'get' method in variable 'response'

response = requests.post(url, headers=headers, data=data)
```

```python
#Check on the response status 

collection=json.loads(response.content)

#Note: The following link has information on the various http response codes that we may encounter while making a request - https://www.w3.org/Protocols/HTTP/HTRESP.html
```

```python
len(collection)
```

```python
df=pd.json_normalize(collection)
```
